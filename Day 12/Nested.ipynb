```python
age = input("Enter your age")
age = int(age)
if age>18:   #outer if
    #Entry
    print("Go to path A and Then towards Target 1")
    if age<=25:    #inner if
        print("You are below 25 or 25, so you can walk and go to Target 2")
    else:    # inner else
        print("You are above 25 so please take a bike and go to Target 2")
else :  #outer else
    #Exit
    print("Go to path B and Exit")
```
