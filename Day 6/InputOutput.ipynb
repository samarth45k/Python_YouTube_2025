### sep vs end (Interview)

### sep is used between two things and it can add a separator between then, default it takes a spaca

### end is used when we want to add anything at the end out output, default it takes a new line (\n)

### New line (\n)

```python
print("Samarth",end = '\n') # new line
print("Kulkarni")
```

### Tab (\t)

```python
print("Samarth",end = '\t\t\t') # tab - horizontal space
print("Kulkarni")
```
